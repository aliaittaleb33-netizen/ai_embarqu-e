```python
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Flatten
from tensorflow.keras.utils import to_categorical
import matplotlib.pyplot as plt
import numpy as np


# TГ©lГ©chargement des donnГ©es MNIST
(x_train, y_train), (x_test, y_test) = tf.keras.datasets.mnist.load_data()

# Normalisation des pixels (les valeurs passent de 0-255 Г  0-1)
x_train = x_train / 255.0
x_test = x_test / 255.0

# Aplatissement des images (28x28 -> 784) comme demandГ© dans le TP
x_train_flat = x_train.reshape(-1, 784)
x_test_flat = x_test.reshape(-1, 784)

# Encodage des labels en one-hot (nГ©cessaire pour la fonction categorical_crossentropy)
y_train_cat = to_categorical(y_train, 10)
y_test_cat = to_categorical(y_test, 10)

print("Format des donnГ©es d'entraГ®nement :", x_train_flat.shape)

```

```python
model_adam = Sequential([
    Dense(128, activation='relu', input_shape=(784,)),
    Dense(10, activation='softmax')
])
model_adam.compile(optimizer='adam', loss='categorical_crossentropy', metrics=['accuracy'])
history_adam = model_adam.fit(x_train_flat, y_train_cat, epochs=10, batch_size=32, validation_split=0.2)
```
